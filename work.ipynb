```python
import pandas as py
dataset = py.read_csv("winemag-data-130k-v2.csv")
dataset.head()
```

```python
import matplotlib.pyplot as plt
dataset["price"].hist(bins=2, color="orange")

plt.xlabel("Points")
plt.ylabel("Frequency")
plt.title("Distribution of Wine Ratings (Points)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))  # wider figure to fit country names
sns.boxplot(x="country", y="price", data=dataset)

plt.xticks(rotation=90)  # rotate country names so they don’t overlap
plt.xlabel("Country")
plt.ylabel("Wine Price (USD)")
plt.title("Wine Price Distribution by Country")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Pick the first 3 countries
top_countries = ["Italy", "Portugal", "US"]
subset = dataset[dataset["country"].isin(top_countries)]

plt.figure(figsize=(3, 4))
sns.boxplot(x="country", y="price", data=subset)

plt.xlabel("Country")
plt.ylabel("Wine Price (USD)")
plt.title("Wine Price Distribution (Italy, Portugal, US)")
plt.show()
```
